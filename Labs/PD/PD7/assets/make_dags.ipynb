```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
options = {
    "font_size": 24,
    "node_size": 1400,
    "node_color": "white",
    "edgecolors": "black",
    "linewidths": 2,
    "width": 5,
}
```

```python
fig, all_axes = plt.subplots(1, 3, figsize = (16, 4))
ax = all_axes.flat
DG = nx.DiGraph()
DG.add_edges_from([("X", "Y"), ("Z", "X"), ("Z", "Y")])
pos = {"X": (0, 0), "Z": (0.5, 0.5), "Y": (1, 0)}
nx.draw(DG, pos, ax = ax[0], with_labels = True, **options)
DG.clear()
DG.add_edges_from([("U", "Z"), ("Z", "X"), ("X", "Y"), ("U", "Y")])
pos = {"X": (0, 0), "Z": (0.5, 0.5), "U": (1, 1), "Y": (2, 0)}
nx.draw(DG, pos, ax = ax[1], **options, with_labels = True)
DG.clear()
DG.add_edges_from([("U", "X"), ("Z", "Y"), ("X", "Y"), ("U", "Z")])
pos = {"X": (0, 0), "Z": (1.5, 0.5), "U": (1, 1), "Y": (2, 0)}
nx.draw(DG, pos, ax = ax[2], with_labels = True, **options)
```

```python
fig, all_axes = plt.subplots(1, 3, figsize = (16, 4))
ax = all_axes.flat
DG = nx.DiGraph()
DG.add_edges_from([("X", "M"), ("Z", "X"), ("Z", "M"), ("M", "Y")])
pos = {"X": (0, 0), "Z": (0.5, 0.5), "M": (1, 0), "Y": (1.5, 0)}
nx.draw(DG, pos, ax = ax[0], with_labels = True, **options)
DG.clear()
DG.add_edges_from([("U", "Z"), ("Z", "X"), ("X", "M"), ("U", "M"), ("M", "Y")])
pos = {"X": (0, 0), "Z": (0.25, 0.25), "U": (0.5, 0.5), "M": (1, 0), "Y": (1.5, 0)}
nx.draw(DG, pos, ax = ax[1], **options, with_labels = True)
DG.clear()
DG.add_edges_from([("U", "X"), ("Z", "M"), ("X", "M"), ("U", "Z"), ("M", "Y")])
pos = {"X": (0, 0), "Z": (0.75, 0.25), "U": (0.5, 0.5), "M": (1, 0), "Y": (1.5, 0)}
nx.draw(DG, pos, ax = ax[2], with_labels = True, **options)
```
